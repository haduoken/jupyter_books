```python
reviews = pd.read_csv("/home/kilox/data/kaggle_data/winemag-data-130k-v2.csv", index_col=0)
# To make pandas use that column for the index (instead of creating a new one from scratch), 
# we can specify an index_col
reviews.shape
reviews.head()
```

```python
# Hence to access the country property of reviews we can use
reviews.country
```

```python
reviews['country'][0]
```

```python
# However, pandas has its own accessor operators, loc and iloc

# To select the first row of data in a DataFrame, we may use the following:

reviews.iloc[0]
```

```python
# Both loc and iloc are row-first, column-second. 
# This is the opposite of what we do in native Python, which is column-first, row-second.
reviews.iloc[:3, 0]
```

```python
# 另外一种是 label_based selection
reviews.loc[0, 'country']
```

```python
# iloc uses the Python stdlib indexing scheme, where the first element of the range is included and the last 
# one excluded. So 0:10 will select entries 0,...,9. loc, meanwhile, indexes inclusively. 
# So 0:10 will select entries 0,...,10.
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

```python
# 这个有10001 行
reviews.loc[0:1000]
# 这个有1000 行
reviews.iloc[0:1000]
```

```python
# 将points进行描述
reviews.points.describe()
```

```python
# 注意不能使用mean带入到lambda里面, 否者会计算超级久
rpm = reviews.points.mean()
reviews.points.map(lambda p: p - rpm)
```
